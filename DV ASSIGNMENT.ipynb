```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Medicaldataset.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Age", y="Result", data=df, palette="viridis")
```

```python
# Plot 1: Histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', kde=True, color='skyblue')
plt.title('Distribution of Patient Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Countplot of Gender
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Scatterplot of Heart rate vs. Systolic blood pressure
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Heart rate', y='Systolic blood pressure', data=df, hue='Result', palette='Set2')
plt.title('Scatterplot of Heart rate vs. Systolic blood pressure')
plt.xlabel('Heart rate')
plt.ylabel('Systolic blood pressure')
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Boxplot of Blood sugar by Result
plt.figure(figsize=(10, 6))
sns.boxplot(x='Result', y='Blood sugar', data=df, palette='muted')
plt.title('Blood sugar distribution by Result')
plt.xlabel('Result')
plt.ylabel('Blood sugar')
plt.tight_layout()
plt.show()
```

```python
# Plot 5: Violinplot of Troponin by Result
plt.figure(figsize=(10, 6))
sns.violinplot(x='Result', y='Troponin', data=df, palette='viridis')
plt.title('Troponin distribution by Result')
plt.xlabel('Result')
plt.ylabel('Troponin')
plt.tight_layout()
plt.show()
```
